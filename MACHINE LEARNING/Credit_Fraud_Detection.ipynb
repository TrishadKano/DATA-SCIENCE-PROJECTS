```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#LOADING THE DATASET

credit_card_data = pd.read_csv(r"C:\Users\trishadp\Downloads\creditcard.csv\creditcard.csv")
```

# THIS DATASET IS HIGHLY IMBALANCED

## 0 > Normal/Legitimate Transactions
## 1 > Fradulent Transactions

```python
# SEPARATING THE DATA FOR ANALYSIS
Legit = credit_card_data[credit_card_data.Class == 0]
Fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(Legit.shape)
print(Fraud.shape)
```

```python
Legit['Amount'].describe()
```

```python
#COMPARING THE VALUES FOR BOTH TRANSACTIONS
credit_card_data.groupby('Class').mean()
```

# DEALING WITH THE IMBALANCED TRANSACTIONS

```python
#USING UNDER SAMPLING
#BUILDING A SAMPLE DATA CONTAINING NORMAL DISTRIBUTION OF FRADULENT AND LEGITIMATE TRANSACTIONS

legit_sample = Legit.sample(n=492)

```

# CONCATENATING THE TWO DATAFRAMES (CLASS)

```python
new_dataset = pd.concat([legit_sample, Fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.Class.value_counts()
```

```python
credit_card_data.groupby('Class').mean()
```

# SPLITTING THE DATASET INTO FEATURES AND TARGETS

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset.Class
```

```python
X
```

```python
Y
```

# SPLITTING THE DATASET INTO TRAINING AND TESTING DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape, Y_train.shape)
```

# MODEL TRAINING

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
model = LogisticRegression(max_iter=1000, solver='lbfgs')
```

```python
# TRAINING THE LOGISTICREGRESSION MODEL WITH TRAINING DATA
model.fit(X_train_scaled, Y_train)
```

# MODEL EVALUATION

```python
# accuracy on training data
X_train_prediction = model.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test_scaled)
training_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', training_data_accuracy)
```
